# Importing the Modules

```python
#numpy,pandas,train_testsplit,matplot.pyplot
import numpy as np
import pandas as pd
```

# Reading the Dataset

```python
dataset = pd.read_csv("insurance_data.csv")
dataset
```

# Sperating Features and Target

```python
x = dataset[["Age"]]
y = dataset["Insurance Bought"]
```

# Building Linear Regression 

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x,y)
```

# Printing the Coefficient , intercept

```python
print(model.coef_)
print(model.intercept_)
```

# Solving using Lineat Regression

```python
coefficient = model.coef_
intercept = model.intercept_

# Calculate the predicted y values using the linear regression equation
#predicted_y = coefficient * X['Age'] + intercept


predicted_y1 = coefficient * 20 + intercept
predicted_y2 = coefficient * 25 + intercept
predicted_y3 = coefficient * 45 + intercept
predicted_y4 = coefficient * 50 + intercept

print(predicted_y1)
print(predicted_y2)
print(predicted_y3)
print(predicted_y4)
```

# Plotting the Y Values predicited against X values

```python

import matplotlib.pyplot as plt

x = [20, 25, 45, 50]
y = [-0.07692308 ,0.11538462 ,0.88461538, 1.07692308]

# Plot the data points
plt.scatter(x, y, color='blue', label='Data Points')

# Plot the line connecting the data points
plt.plot(x, y, color='red', linestyle='-', linewidth=2, label='Line')

# Add labels and title
plt.xlabel('Age')
plt.ylabel('Predicted Insurance Bought')
plt.title('Plot of Predicted Values')
plt.legend()

# Show the plot
plt.show()
```

# Building a Sigmoid Function

```python
import math
def sigmoid(x):
  return 1 / (1 + math.exp(-x))
```

```python
y1 = sigmoid(predicted_y1)
y2 = sigmoid(predicted_y2)
y3 = sigmoid(predicted_y3)
y4 = sigmoid(predicted_y4)
print(y1)
print(y2)
print(y3)
print(y4)
```
