# Creating a ANN Model

```python
# Sequential means will accpet every layer as one element
# Dense means first layer of neurons is connected to next layer
# input shpe = 784 , output = 10 , activation function sigmoid
#sparse means the output is an integer array
# categorical if output is onehot encoded
```

```python
import matplotlib.image as mpimg
img = mpimg.imread('NNHandwriting.jpg')

# Display the image
plt.imshow(img)
plt.axis('off') # Hide the axis
plt.show()
```

```python
model = keras.Sequential([
    keras.layers.Dense(10, input_shape=(784,), activation='sigmoid') 
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
              
model.fit(X_train_flattened, y_train, epochs=5) # training

```

```python
#accuracy 89%
```

# Stop move to with scalling

```python

model.evaluate(X_test_flattened, y_test)
```

```python
#printing the image of X_test at index 0
```

```python
plt.matshow(X_test[0])
```

```python
Y_pred = model.predict(X_test_flattened)
```

```python
Y_pred[0]
```

```python
np.argmax(Y_pred[0])# which ouptut neuron has high value 
```

```python
y_predicted_labels = [np.argmax(i) for i in Y_pred] # creating labels
y_predicted_labels[:5] # first 5 predictions
```

```python
cm = tf.math.confusion_matrix(labels=y_test,predictions=y_predicted_labels)
cm
```

```python
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

# Adding Hidden Layer

```python
model = keras.Sequential([
   # keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(100, activation='relu'), # 100 neurons
    keras.layers.Dense(10, activation='sigmoid')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(X_train_flattened, y_train, epochs=10)
```

```python
#Lets Evalute
```
