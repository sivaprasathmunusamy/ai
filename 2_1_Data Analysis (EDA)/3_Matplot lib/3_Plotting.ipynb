# Scatter Plot

```python
# Importing the necessary library
import matplotlib.pyplot as plt

# Step 1: Create data
temp = [30, 35, 40, 45, 50]  # x-coordinates
icecreamsales = [27, 34, 31, 45, 53]  # y-coordinates (y = 2x)


# Step 2: Create a scatter plot
plt.scatter(temp, icecreamsales, color='red', label='Data points')  # Scatter plot with blue points
plt.legend()
```
