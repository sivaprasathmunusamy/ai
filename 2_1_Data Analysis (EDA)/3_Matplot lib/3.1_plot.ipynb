```python
# Importing the necessary library
import matplotlib.pyplot as plt

# Step 1: Create data
x = [1, 2, 3, 4, 5]  # x-coordinates
y = [2, 4, 6, 8, 10]  # y-coordinates (y = 2x)

# Step 2: Create a scatter plot
plt.plot(x, y, color='blue', label='Data points')  # Scatter plot with blue points

# Step 3: Customize the plot
plt.xlabel('x values')  # Label for the x-axis
plt.ylabel('y values')  # Label for the y-axis
plt.title('Simple Scatter Plot')  # Title of the plot
plt.legend()  # Display the legend

# Step 4: Display the plot
plt.show()
```
