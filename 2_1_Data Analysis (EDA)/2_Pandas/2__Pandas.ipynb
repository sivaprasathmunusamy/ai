* Any other form of observational / statistical data sets. The data actually need not be labeled at all to be placed into a pandas data structure

### Pivot 

__Pivot__ is used to transform or reshape a dataframe.

Let's import a csv file.

To have `date` in x-axis and `city` in y-axis, __pivot__ is used as follows.

### Pivot Table

__Pivot Table__ is used to summarize and aggregate data in a dataframe.

Let's import a csv file.

```python
df7 = pd.read_csv("weather2.csv")
df7
```

The above dataframe shows the `temperature` and `humidity` in the morning and evening of two cities for couple of days.

Now, to create a table which contains `average temperature` throughout the day, __pivot table__ can be used as follows.

```python
df7.pivot_table(index = 'city', columns = 'date')
```

`average` or `mean` is the default attribute for pivot_table. However, if any other operation is needed, it should be explicitly mentioned as `aggfunc` attribute.

For example, if we need to `sum` it instead of taking the `mean` of the values,

```python
df7.pivot_table(index = 'city', columns = 'date', aggfunc = 'sum')
```

If the `mean` value is needed alongside the actual values, `margins` attribute shall be used as follows.

```python
df7.pivot_table(index = 'city', columns = 'date', margins = True)
```

So, `margins` gives the `mean` values both row-wise and column-wise (under `All`) along with the actual values.

If other operation is needed (other than `mean`), shall be explicitely mentioned.

```python
df7.pivot_table(index = 'city', columns = 'date', aggfunc = 'sum', margins = True)
```

### Handling Missing Data

Many a times, data will be incomplete. For example, observe the following data.

```python
df9 = pd.read_csv("weather4.csv")
df9
```

The above data is missing many information represented by `NaN` and even the data for 2nd and 3rd of August is missing.

To handle this, pandas have different methods.

Before going further, in the above dataframe, each data in the `date` column is a `string` type, as it is imported from a csv file.

Let's check it.

```python
type(df9.date[0])
```

But, for the purpose of handling missing values with dates using different methods in pandas, this shall be changed to a date-time format.

This is done as follows.

```python
df9 = pd.read_csv("weather4.csv", parse_dates = ['date'])
df9
```

Now, checking the type of the `date` column,

```python
type(df9.date[0])
```

Now, the type is changed to `Timestamp`, which is a date-time format.

Before handling the missing values, let's make the `date` column as the index of the dataframe.

```python
df9.set_index('date', inplace = True)
df9
```

Making `inplace` as `True` modifies the existing dataframe, instead of returning a new one.

There are differnt methods to handle missing data. Let's study 3 of them here.

1. fillna
2. interpolate
3. dropna

__fillna__

`fillna` method is used to fill the values in different ways.

For example, let's fill the missing values with `zeros'.

```python
new_df = df9.fillna(0)
new_df
```

Every `NaN` value is replaced by a `zero` now.

However, filling all the missing values with zero might not make much sense. For example, filling the missing value in the `event` column with a zero is meaningless.

So, each column can be specifically filled with different values using a `dictionary` as follows.

```python
new_df = df9.fillna({
    'temperature' : 0,
    'windspeed' : 0,
    'event' : 'no event'
})

new_df
```

Missing values in `temperature` and `windspeed` columns are replaced with zeros and in `event` column, replaced with `no event`.

Filling the missing values for temperature and windspeed as zero doesn't make much sense. But, filling them with the mean value also won't work.

So, the value of the previous entry can be used to fill the missing value.

This is done using `forwardfill (ffill)` of `method` attribute as follows.

```python
new_df = df9.fillna(method = 'ffill')
new_df
```

Now, the missing data is filled up using the previous value.

Instead of using the previous value, missing values can be filled using the sucessive value.

This is done using `backwardfill (bfill)` of `method` attribute as follows.

```python
new_df = df9.fillna(method = 'bfill')
new_df
```

Now, the missing data is filled up using the sucessive value.

For both ffill and bfill, filling was row-wise. To make the filling process column-wise, `axis` attribute is used as follows.

```python
new_df = df9.fillna(method = 'bfill', axis = 'columns')
new_df
```

Considering the ffill again,

```python
new_df = df9.fillna(method = 'ffill')
new_df
```

```python
df9
```

Here, missing values are filled with the previous values till the next actual value. That is, for both 8th and 9th of August, missing value was filled with '33', till 10th of August which has the actual value.

To, restrict the filling process to the requirement instead of complete filling, `limit` attribute is used as follows.

```python
new_df = df9.fillna(method = 'ffill', limit = 1)
new_df
```

Even though the missing values are filled with the previous values or sucessive values, it may not make sense sometimes.

For example, temperature on 1st of August was 33 and on 5th was 29. Then, logically temperature on 4th shall be in between 33 and 29 (however, need not be true). 

So, to fill these intermediate values instead of filling the previous or successive values, __interpolate__ method is used.

__Interpolate__

```python
new_df = df9.interpolate()
new_df
```

Now, the guess for temperature on 4th is `31`, instead of `33` which is comparitively better.

This is a linear interpolation. By default, `method` attribute takes `linear` for interpolation.

But, still the guess of `31` is not that accurate as the value for 4th shall be nearer to 5th, instead of 1st.

To rectify this, `time` attribute is used.

Now, let's change the `method` attribute to `time` as follows.

```python
new_df = df9.interpolate(method = 'time')
new_df
```

Now, the guess for 4th is '30', which is a better guess.

Instead of filling `NaN` with some values, it could be dropped off the dataframe. To do this, __dropna__ method is used.

__Dropna__

```python
new_df = df9.dropna()
new_df
```

This has clearly dropped all the rows which had `NaN` and has retained the rows which had all the valid values.

To retain the rows with definite number of `NaN`, `how` attribute is used.

For example, to drop the rows which have every entry `NaN`,

```python
new_df = df9.dropna(how = 'all')
new_df
```

To retain the rows which atleast have one non - NaN value, `thresh (threshold)` attribute is used as follows.

```python
new_df = df9.dropna(thresh = 1)
new_df
```

Apparently, 'how = all' and 'thresh = 1' are same as both of them won't retain the row with no valid values.

To retain the rows with atleast two valid values, threshold shall be set to 2.

```python
new_df = df9.dropna(thresh = 2)
new_df
```

### Concatenate

__Concatenate__ (used as `Concat`), is an operataion used to join one or more dataframes.

Let's create a dataframe as follows.

```python
india_weather = pd.DataFrame({
    'city' : ['Bengaluru', 'Pune', 'Chandigarh'],
    'temperature' : [33, 46, 31],
    'humidity' : [81, 62, 79]
})

india_weather
```

Similarly, creating another dataframe as follows.

```python
us_weather = pd.DataFrame({
    'city' : ['chicago', 'newyork', 'orlando'],
    'temperature' : [15, 23, 37],
    'humidity' : [67, 64, 76]
})

us_weather
```

The above two dataframes can be joined to get a single dataframe using `concat` as follows.

```python
df10 = pd.concat([india_weather, us_weather])
df10
```

But, the index is retained as it is. However, the index can be made continuous using `ignore_index` argument as follows.

```python
df10 = pd.concat([india_weather, us_weather], ignore_index = True)
df10
```

`Keys` argument can be used to associate with each dataframe.

```python
df10 = pd.concat([india_weather, us_weather], keys = ['india', 'us'])
df10
```

These additional indices can be used in data selection as follows.

```python
df10.loc['india']
```

```python
df10.loc['us']
```

To join the datafram column-wise, instead of row-wise, `axis` argument is used.

Consider the following dataframes.

```python
temperature_df = pd.DataFrame({
    'city' : ['Bengaluru', 'Pune', 'Chandigarh'],
    'temperature' : [33, 46, 31],
})
temperature_df
```

```python
windspeed_df = pd.DataFrame({
    'city' : ['Bengaluru', 'Pune', 'Chandigarh'],
    'windspeed' : [8, 13, 11],
})
windspeed_df
```

If basic `concat` is used, dataframes will be joined row-wise as follows.

```python
df11 = pd.concat([temperature_df, windspeed_df])
df11
```

To join the dataframes column-wise, `axis` argument is used as follows.

```python
df11 = pd.concat([temperature_df, windspeed_df], axis = 1)
df11
```

Due to an imperfect data, if we have windspeed values missed and jumbled, where values Chandigarh are missing and order of the cities are jumbled as shown below,

```python
windspeed_df = pd.DataFrame({
    'city' : ['Pune', 'Bengaluru'],
    'windspeed' : [8, 13],
})
windspeed_df
```

If we concatenate now,

```python
df11 = pd.concat([temperature_df, windspeed_df], axis = 1)
df11
```

It just appended them row-wise and is incorrect. To rectify that, `index` argument is used as follows.

```python
temperature_df = pd.DataFrame({
    'city' : ['Bengaluru', 'Pune', 'Chandigarh'],
    'temperature' : [33, 46, 31],
}, index = [0, 1, 2])
temperature_df
```

```python
windspeed_df = pd.DataFrame({
    'city' : ['Pune', 'Bengaluru'],
    'windspeed' : [8, 13],
}, index = [1, 0])
windspeed_df
```

Now, the indices are made identical explicitely and the concatenation will be proper.

```python
df11 = pd.concat([temperature_df, windspeed_df], axis = 1)
df11
```

So, `index` argument is used to allign rows from different dataframes during concatenation.

Consider the following series `s1` which represents the general weather condition.

```python
s1 = pd.Series(["Rainy", "Dry", "Humid"], name = 'weather')
s1
```

Our temperature dataframe is,

### Merge

__Merge__ joins the dataframes in a proper way, without `index` argument used explicitely.

Consider the following dataframes.

```python
df13 = pd.DataFrame({
    'city' : ['newyork', 'chicago', 'orlando'],
    'temperature' : [15, 23, 37],
})

df13
```

```python
df14 = pd.DataFrame({
    'city' : ['chicago', 'newyork', 'orlando'],
    'humidity' : [67, 64, 76]
})

df14
```

To get a dataframe which contains both the `temperature` and `humidity` data, `merge` is used as follows.

```python
df15 =  pd.merge(df13, df14)
df15
```

It is clearly evident that, even though city names aren't in the order, the merged dataframe has correct values.

That is, `merge` hasn't considered the index blindly. The resultant dataframe has the correct values for both `temperature` and `humidity`.

The same operation can be done by explicitely mentioning the column name with respect to which, the dataframes shall be merged as follows.

```python
df15 =  pd.merge(df13, df14, on = 'city')
df15
```

If the number of cities are different or the cities themselves are different in these dataframes, `merge` works.

Consider the following dataframes.

```python
df13 = pd.DataFrame({
    'city' : ['newyork', 'chicago', 'orlando', 'Tampa'],
    'temperature' : [15, 23, 37, 33],
})

df13
```

```python
df14 = pd.DataFrame({
    'city' : ['chicago', 'newyork', 'sandiego'],
    'humidity' : [67, 64, 72]
})

df14
```

```python
df15 =  pd.merge(df13, df14, on = 'city')
df15
```

There are different ways and default is merging with only the common values (intersection) as shown above. 

The type of merging can be decided using the argument `how`. By default, `how` is `inner`. 

```python
df15 =  pd.merge(df13, df14, on = 'city', how = 'inner')
df15
```

Instead of intersection, if union is needed, `outer` is used.

```python
df15 =  pd.merge(df13, df14, on = 'city', how = 'outer')
df15
```

Similarly, `left` and `right` can be used for the argument `how`.

```python
df15 =  pd.merge(df13, df14, on = 'city', how = 'left')
df15
```

```python
df15 =  pd.merge(df13, df14, on = 'city', how = 'right')
df15
```

`Right` and `Left` are with respect to the positioning of the dataframe names in `merge` function.
