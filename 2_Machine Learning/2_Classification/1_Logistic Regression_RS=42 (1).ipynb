# Step : 1 importing the Libraries

```python
import numpy as np
import pandas as pd
```

# Step 2 : Reading the Dataset

```python
dataset = pd.read_csv("cancer_dataset.csv")
```

```python
dataset.head(7)
```

```python
x = dataset.iloc[:,2:32]
y = dataset.iloc[:,1]

print(y.head(3))
print("********")
print(x.ndim)
print(y.ndim)
```

# Step 3 : Seperating the Features and Target

```python
x = dataset[["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"]]
```

```python
x.head()
```

```python
y = dataset["Species"]
```

```python
y.head()
```

# Step 4 : Splitting the dataset into training set and test set

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.8 ,random_state=-42)
```

# Step 5: Creating the Model

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

# Step 6 : Training the Model

```python
model.fit(x_train,y_train)
```

# Step 7 : Testing the Model

```python
model.predict([[5.1,3.5	,1.4,0.2]])
```

```python
y_pred = model.predict(x_test)
```

# Step 8: Checking the Accuracy Score

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test,y_pred)
print(acc*100)
```

# Step 9 : Create Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
print(cm)
```
