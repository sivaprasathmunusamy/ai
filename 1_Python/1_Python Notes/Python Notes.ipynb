# List Comphrehension

```python
a = [2,3,4,5,6]
b = []

for i in a:
    x = i*2
    b.append(x)
print(b)
```

```python
a = [2,3,4,5,6]
b = []

for i in a:
    if i>4:
        x = i*2
        b.append(x)
print(b)
```

```python
a = [2,3,4,5,6]
b = [i*2 for i in a]
print(b)
```

```python
a = [2,3,4,5,6]
b = [i*2 for i in a if i>4]
print(b)
```

# Dictionary Comphrension

```python
a = {"A":10 ,"B":20 , "C":30 , "E":40 }
b = {i:j+100 for i,j in a.items()}
print(b)
```

```python
a = {"A":10 ,"B":20 , "C":30 , "E":40 }
b = {i:j+100 for i,j in a.items() if j>20}
print(b)
```
